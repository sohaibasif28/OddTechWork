```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from xgboost import XGBClassifier
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Full_Finalized_numeric_data_for_training.csv')
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
target = df['Label']
```

```python
scaler = StandardScaler()
features = scaler.fit_transform(df[['cityname', 'persongender','personexactage','investment', 'CreditRating', 'hour']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.1)
```

```python
model = XGBClassifier(random_state=42,
                      learning_rate = 0.01,
                      subsample = 0.8,
                      n_estimators=1000,
                      max_depth=11,
                      colsample_bytree = 0.8,
                      gamma=1)
model_fit = model.fit(X_train, y_train)
```

```python
model_predict=model.predict(X_test)
```

```python
model_predict
```

```python
print('Forest Train accuracy %s' % model.score(X_train, y_train))
```

```python
print('Trees Test accuracy %s' % accuracy_score(model_predict, y_test)) 
```
