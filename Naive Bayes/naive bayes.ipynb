```python
import pandas as pd
```

```python
df = pd.read_csv('Full_Finalized_numeric_data_for_training.csv')
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
target = df['Label']
```

```python
features = df[['state', 'persongender', 'investment','PresenceOfCreditCard', 'CreditRating']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```
